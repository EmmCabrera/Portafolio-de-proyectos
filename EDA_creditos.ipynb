```python
import polars as pl
```

```python
# Cargar set de datos
df_trabajo = pl.read_csv("application_train.csv")
```

```python
df_trabajo.head()
```

```python
df_trabajo.describe()
```

```python
#porcentaje de nulos
null_porcentaje = (df_trabajo.null_count() / df_trabajo.height) * 100
print(null_porcentaje)
```

```python
# Dataframe con columnas cuyos nulos es menor al 20%
columns_a_mantener = [col for col in null_porcentaje.columns if null_porcentaje[0, col] <= 0.20]
print(columns_a_mantener)
df_filtrado = df_trabajo.select(columns_a_mantener)
df_filtrado.describe()
```

```python
# Identificación de columnas con nulos y tipo de datos previa a la imputación de datos
nulos_df = df_filtrado.null_count()
columnas_con_nulos_y_tipos = {
    col: df_filtrado.schema[col]
    for col in df_filtrado.columns
    if nulos_df[0, col] > 0}
print(columnas_con_nulos_y_tipos)
```

```python
# Imputa valores nulos en columnas numéricas y de texto en una sola operación
df_imputado = df_filtrado.with_columns(
    # Para columnas numéricas, rellena los nulos con la media
    pl.col(pl.Float64).fill_null(pl.col(pl.Float64).mean()),
    pl.col(pl.Int64).fill_null(pl.col(pl.Int64).mean()),

    # Para columnas de texto, rellena los nulos con "sin información"
    pl.col(pl.Utf8).fill_null(pl.lit("sin información"))
)

print(df_imputado.describe())
```

```python
#elaboración del PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
# Columnas númericas para el PCA
columnas_float = {
    col: dtype
    for col, dtype in df_imputado.schema.items()
    if dtype == pl.Float64
}
print(df_imputado.schema)

features = list(columnas_float.keys())
X_polars = df_imputado.select(features)
X = X_polars.to_numpy()

# Estandarizar y realizar el PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA()
pca.fit(X_scaled)

# Obtener y mostrar los coeficientes (loadings)
loadings = pca.components_
print("Coeficientes de los Componentes Principales:\n", loadings)

# Dataframe
loadings_df = pl.DataFrame(
    loadings,
    schema=[f"PC{i+1}" for i in range(loadings.shape[0])],
)
loadings_df = loadings_df.with_columns(pl.Series("Variable", features)).transpose(include_header=True)
pl.Config.set_tbl_rows(None)
pl.Config.set_tbl_cols(None)
print("\nDataFrame de Coeficientes:\n", loadings_df)
```

```python
import polars as pl
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Seleccionar las columnas numéricas para el PCA
columnas_float = {
    col: dtype
    for col, dtype in df_imputado.schema.items()
    if dtype == pl.Float64
}
features = list(columnas_float.keys())
X_polars = df_imputado.select(features)
X = X_polars.to_numpy()

# Estandarizar los datos y realiza el PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA()
principal_components = pca.fit_transform(X_scaled)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# Gráfico de Dispersión
ax1.scatter(principal_components[:, 0], principal_components[:, 1])
ax1.set_xlabel('Principal Component 1')
ax1.set_ylabel('Principal Component 2')
ax1.set_title('Gráfico de Dispersión de los Componentes Principales')
ax1.grid(True, linestyle='--', alpha=0.6)

# Gráfico de Codo
explained_variance = pca.explained_variance_ratio_
componentes = np.arange(1, len(explained_variance) + 1)
varianza_acumulativa = np.cumsum(explained_variance)

ax2.bar(componentes, explained_variance, alpha=0.7, color='b', label='Varianza Explicada por Componente')
ax2.plot(componentes, varianza_acumulativa, marker='o', color='r', linestyle='-', label='Varianza Acumulada')

ax2.set_xlabel('Número de Componente Principal')
ax2.set_ylabel('Varianza Explicada')
ax2.set_title('Gráfico de Codo (Scree Plot)')
ax2.set_xticks(componentes)
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend()

plt.tight_layout()
plt.show()
```

Componente Principal 1 (PC1)
PC1 está fuertemente influenciado por las variables relacionadas con el crédito y el monto de los bienes, la columnas que más influyen son AMT_CREDIT, AMT_ANNUITY y AMT_GOODS_PRICE, ya que tienen los coeficientes (loadings) más altos en valor absoluto:

AMT_GOODS_PRICE (0.581)

AMT_CREDIT (0.579)

AMT_ANNUITY (0.534)

De igual forma se puede interpretar que este componente representa un factor general de "tamaño del préstamo"; por ende, un valor alto en PC1 indicaría un cliente con un crédito grande, una anualidad alta y un precio de bienes elevado.

Componente Principal 2 (PC2)
PC2 está principalmente definido por variables de tiempo y estructura familiar,las columnas con mayor influencia son DAYS_REGISTRATION y CNT_FAM_MEMBERS.

DAYS_REGISTRATION (0.691)

CNT_FAM_MEMBERS (0.619)

Este componenteesta relacionado con la estabilidad y el historial del cliente, al combinar la antigüedad de su registro con el tamaño de su familia.
